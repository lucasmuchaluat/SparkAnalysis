```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
df1 = pd.read_csv("ResultadoPalavra1.csv")
df2 = pd.read_csv("ResultadoPalavra2.csv")
dfinter = pd.read_csv("ResultadoIntersect.csv")
#local
df_inter_local = pd.read_csv("ResultadoIntersectLocal.csv")
df_palavra1_local = pd.read_csv("ResultadoPalavra1Local.csv")
df_palavra2_local = pd.read_csv("ResultadoPalavra2Local.csv")
```

```python
df1.head()
```

```python
def Convert(df):
    di = {}
    for index in range(len(df)): 
        di[df["Palavra"][index]] = df["Relevância"][index] 
    return di
```

```python
wordcloud = WordCloud(width=2000, height=1000, background_color="white")
wordcloud.generate_from_frequencies(frequencies=Convert(df1))
plt.figure(figsize = (15,15))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("BOULOS")
plt.show()
```

```python
wordcloud = WordCloud(width=2000, height=1000, background_color="white")
wordcloud.generate_from_frequencies(frequencies=Convert(df2))
plt.figure(figsize = (15,15))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("COVAS")
plt.show()
```

```python
wordcloud = WordCloud(width=2000, height=1000, background_color="white")
wordcloud.generate_from_frequencies(frequencies=Convert(dfinter))
plt.figure(figsize = (15,15))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("INTER")
plt.show()
```
